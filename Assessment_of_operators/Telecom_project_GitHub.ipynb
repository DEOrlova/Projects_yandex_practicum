```python
td = pd.read_csv('telecom_dataset.csv')
tc = pd.read_csv('telecom_clients.csv')
```

```python
# функция для базовой оценки структуры данных
def data_preprocessing(df):
    display(df.head())
    print('*'*50)
    df.info()
    print('*'*50)
    print('Доля пропусков: ', df.isna().mean().sort_values(ascending=False), sep='\n')
    print('*'*50)
    print('Полных дубликатов: ', df.duplicated().sum())
    print('*'*50)
    print(f'Исходный дф содержит: {df.shape[0]} строк')
    print('*'*50)
```

Изучим подробнее аномалии каждого столбца. Для этого напишем функцию.

```python
def anomalies(df, name_column):
    '''
       Ф-я для исследования аномалий в столбце
       
       Принимает название дф и столбца.
       Возвращает значение перцентилей,
       статистические данные
       и 3 графика:
          диаграмму рассеивания,
          выбросы боксплота,
          боксплот без выбросов.
    '''    
    
    fig, ax = plt.subplots(1, 3, figsize=(10, 5))
    plt.suptitle(f'Распределение значений в столбце {name_column}')
    ax[0].set_title('Диаграмма рассеивания')
    ax[1].set_title('Выбросы')
    ax[2].set_title('Боксплот без выбросов')
    
    sns.scatterplot(x='operator_id', y=name_column, data=df, ax=ax[0])
    df.boxplot(column=name_column, ax=ax[1])
    df.boxplot(column=name_column, showfliers=False, ax=ax[2])
    
    perc = np.percentile(df[name_column], [90, 95, 99])
    
    return (plt.show(),
    print(f'Значение 90, 95 и 99 перцентилей: {perc}'),
    print('*'*50),
    print(df[name_column].describe()),
    print('*'*50))
```

### Определение метрик
<a id="metric"></a>

Для определения метрик эффективности операторов, попробуем классифицировать их работу.

В идеале - нам нужно разделить операторов по специфике работы на исходящие и входящие звонки. Но в процессе деления мы увидим, что есть операторы, которые работают и на вход и на исход. Поэтому будет три группы операторов и три группы метрик оценки эффективности:

1) только исходящие звонки

2) только входящие звонки

3) смешанные звонки

Агрегируем данные по столбцу `direction`, чтобы разделить опреаторов на категории.

При этом у нас остается разделение на внутренние и внешние звонки. Суммируем их вместе, т.к. операторы могут переводить звонки друг на друга, это тоже часть работы. При этом два последних столбца со средними значениями возьмем так же по среднему.

Столбец `is_missed_call` оставим только со значением не пропущенных.

```python
# группируем данные, чтобы определить операторов, кто совершает звонки одного типа
df_group = df.groupby(['operator_id']).agg({'direction' : 'nunique'}).query('direction == 1')

# отбираем тип исходящих
df_spec_out = df.query("(direction == 'out')&(operator_id in @df_group.index)&(is_missed_call == False)")\
                .groupby(['date', 'operator_id'], as_index=False)\
                .agg({'calls_count': 'sum',
                      'call_duration': 'sum',
                      'total_call_duration': 'sum',
                      'avg_one_duration': 'mean',
                      'avg_wait_duration': 'mean'})

# отбираем тип входящих
df_spec_in = df.query("(direction == 'in')&(operator_id in @df_group.index)&(is_missed_call == False)")\
                .groupby(['date', 'operator_id'], as_index=False)\
                .agg({'calls_count': 'sum',
                      'call_duration': 'sum',
                      'total_call_duration': 'sum',
                      'avg_one_duration': 'mean',
                      'avg_wait_duration': 'mean'})

# отбираем смешанный тип
df_mix = df.query("(operator_id not in @df_group.index)&(is_missed_call == False)")\
            .groupby(['date', 'operator_id'], as_index=False)\
                .agg({'calls_count': 'sum',
                      'call_duration': 'sum',
                      'total_call_duration': 'sum',
                      'avg_one_duration': 'mean',
                      'avg_wait_duration': 'mean'})
```

### Оценка операторов - вариант1
По выявленным показателям оценим операторов КЦ. Для этого напишем функцию, которая проставит баллы операторам в новом столбце.

```python
def estimator(column, num1, num2):
    '''
       Ф-я для присвоения категории оператору
       
       Принимает столбец.
       Возвращает категорию из расчета по 50 и 75 перцентилю распределения значений столбца:
       0 - не эффективен,
       1 - эффективен,
       2 - сверхэффективен.
    '''
    
    try:
        if 0 <= column < num1:
            return 0
        elif num1 <= column < num2:
            return 1
        elif num2 <= column:
            return 2
        
    except:
        pass
```

С помощью вложенного цикла создадим три новых столбца с оценками в каждом дф.

```python
for d in [df_spec_out, df_mix]:
    for c in ['calls_count', 'call_duration', 'avg_wait_duration']:
        d['r_'+c] = d[c].apply(lambda x: estimator(x, np.percentile(d[c], 50), np.percentile(d[c], 75)))
    display(d.head())
```

И в последнем столбце по длительности ожидания поменяем местами 0 и 2.

```python
for d in [df_spec_out, df_mix]:
    for c in ['r_avg_wait_duration']:
        d[c] = d[c].map({0: 2, 2: 0, 1:1})
    display(d.head())
```

Создадим столбец со средней дневной оценкой оператора по трем показателям.

```python
# скомбинируем операторов с рейтингами в один дф
df_rating = pd.concat([df_spec_out, df_mix])

# и создадим столбец со средней дневной оценкой оператора по трем показателям
df_rating['avg_rating_day'] = (df_rating[['r_calls_count','r_call_duration','r_avg_wait_duration']]
                                        .sum(axis=1)/3).astype('int64')
df_rating
```

Мы получили "оценку" эффективности операторов двух групп КЦ в разрезе дня.

Теперь для оценки операторов за весь период, суммируем оценки каждого оператора и разделим на количество отработанных им дней. Получим среднюю оценку оператора за весь период. Даже если оператор в какой-то день провалил план, но в другой сделал больше, он компенсирует свой рейтинг.

### Оценка операторов - вариант2
Попробуем еще одну методику оценки - выберем три критерия (наиболее вероятные в работе любого КЦ) и применим оценку с расстановкой весов.

Допустим, значимыми являются три метрики:
- количество совершенных звонков,
- длительность разговора без ожидания
- и среднее время ожидания клиента (инвертированная).

Считаем, что наибольший вес имеет количество звонков, затем общая длительность разговоров и совсем маленький вес у ожидания клиента на линии (т.к. это метрика в меньшей степени зависит от оператора, если только он не резко выделяется из всех остальных).

Возьмем разделенные и агрегированные дф и подготовим их для введения весовой оценки.

Датафреймы берем все те же, код скопируем заново, чтобы обнулить все изменения в первом варианте оценки и немного скорректируем названия.

```python
# группируем данные, чтобы определить операторов, кто совершает звонки одного типа
df_group = df.groupby(['operator_id']).agg({'direction' : 'nunique'}).query('direction == 1')

# отбираем тип исходящих
spec_out = df.query("(direction == 'out')&(operator_id in @df_group.index)&(is_missed_call == False)")\
                .groupby(['date', 'operator_id'], as_index=False)\
                .agg({'calls_count': 'sum',
                      'call_duration': 'sum',
                      'total_call_duration': 'sum',
                      'avg_one_duration': 'mean',
                      'avg_wait_duration': 'mean'})

# отбираем тип входящих
spec_in = df.query("(direction == 'in')&(operator_id in @df_group.index)&(is_missed_call == False)")\
                .groupby(['date', 'operator_id'], as_index=False)\
                .agg({'calls_count': 'sum',
                      'call_duration': 'sum',
                      'total_call_duration': 'sum',
                      'avg_one_duration': 'mean',
                      'avg_wait_duration': 'mean'})

# отбираем смешанный тип
mix = df.query("(operator_id not in @df_group.index)&(is_missed_call == False)")\
            .groupby(['date', 'operator_id'], as_index=False)\
                .agg({'calls_count': 'sum',
                      'call_duration': 'sum',
                      'total_call_duration': 'sum',
                      'avg_one_duration': 'mean',
                      'avg_wait_duration': 'mean'})
```

Напишем функцию для простановки рейтингов исходя из весов показателей. Т.к. распределения значений в группах отличаются, каждый дф будем пропускать в функцию отдельно, чтобы нормализация была в ряду свойсвенных группе значений.

Ф-я может быть и более универсальной, но в нашем случае отличается только дф, хотя давайте предположим изменение весов.

```python
def in_rating(df, wcc=0.5, wcd=0.4, wwd=0.1):
    # Преобразование значений для учета лучшего показателя
    df['inverse_wait_duration'] = 1 / df['avg_wait_duration']

    # нормализуем данные в наших столбцах методом ML из библиотеки sklearn
    # создаём объект класса scaler
    scaler = MinMaxScaler()

    # отбираем столбцы для преобразования
    columns_to_normalize = ['calls_count', 'call_duration', 'avg_wait_duration']

    # преобразуем набор данных
    df[['norm_calls_count', 'norm_call_duration', 'norm_avg_wait_duration']] = scaler.fit_transform(df[columns_to_normalize])

    # Зададим значения весов

    # количество звонков
    wcc = 0.5
    # длительность звонков
    wcd = 0.4
    # средняя длительность ожидания
    wwd = 0.1

    #пропишем формулу оценки эффективности операторов
    df['rating_day'] = (wcc * df['norm_calls_count'] +
                    wcd * df['norm_call_duration'] +
                    wwd * df['norm_avg_wait_duration']).round(1)
    return df
```

```python
def result(df):
    min = df.rating_day.min()
    max = df.rating_day.max()
    cnt = df.rating_day.value_counts()
    return print(f'Диапазон рейтингов от {min} до {max}'), print('\nРаспределение рейтингов', cnt, sep='\n')
```

```python
rating_in = in_rating(spec_in)
result(rating_in)
```

```python
rating_mix = in_rating(mix)
result(rating_mix)
```

С учетом полученных результатов для объединения дф и категоризации этого столбца, будем считать так:

- 0.3-0.4 = 1 - эффективен
- 0.5-0.6 = 2 - сверхэффективен.

## Статистические гипотезы
На базе наших данных проверим две статистические гипотезы.

Т.к. в данных есть КЦ трех типов звонков и у них априори разные показатели по звонкам, для проверки гипотез возьмем **третью категорию операторов `mix`**, которые работают со звонками обоих типов.

### Гипотеза 1:
$\begin{equation*}
 \begin{cases}
   H_0 :\text{Средняя длительность отвеченных звонков для входящих и исходящих вызовов одинакова}\\
   H_1 :\text{Средняя длительность отвеченных звонков для входящих и исходящих вызовов различна}
 \end{cases}
\end{equation*}$

Уровень значимости возьмем наиболее универсальный: $\alpha = 0.05$

```python
test_in = df.query("(operator_id not in @df_group.index)&(is_missed_call == False)&(direction == 'in')")['call_duration']
test_out = df.query("(operator_id not in @df_group.index)&(is_missed_call == False)&(direction == 'out')")['call_duration']
print(test_in.shape[0], test_out.shape[0])
```

Напишем функцию для проверки на нормальность распределения с возможностью менять значение $\alpha$:

```python
def for_shapiro(sample, name, alpha=0.05):
    
    stat, p_value = shapiro(sample)
    alpha = alpha

    print('Статистика теста:', stat)
    print('p-value:', p_value)

    if p_value > alpha:
        print(f'Выборка {name} имеет нормальное распределение')
    else:
        print(f'Выборка {name} не имеет нормального распределения')
```

```python
for_shapiro(test_in, 'test_in')
```

```python
for_shapiro(test_out, 'test_out')
```

Выборки не имеют нормального распределения, поэтому лучше использовать параметрический тест Манн-Уитни (он не требует предположения о нормальности распределения данных, в отличие от t-теста. Это делает его более устойчивым к выбросам и аномальным значениям данных. Поэтому он лучше подходит для нашего анализа).

Напишем функцию для проведения теста Манн-Уитни:

```python
def for_test(sample1, sample2, alpha=0.05):
    
    stat, p_value = mannwhitneyu(sample1, sample2, alternative='two-sided')
    alpha = alpha

    print('p-value:', p_value)

    if p_value > alpha:
        print('Отвергаем нулевую гипотезу: разница статистически значима')
    else:
        print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')
```

```python
for_test(test_in, test_out)
```

p-value очень маленькое, однозначно не можем отвергнуть нулевую гипотезу, а значит, не можем утверждать, что длительность входящих и исходящих звонков отличается.

### Гипотеза 2:
$\begin{equation*}
 \begin{cases}
   H_0 :\text{Среднее количество отвеченных внешних и внутренних вызовов одинаково}\\
   H_1 :\text{Среднее количество отвеченных внешних и внутренних вызовов различно}
 \end{cases}
\end{equation*}$

Уровень значимости возьмем наиболее универсальный: $\alpha = 0.05$

```python
internal = df.query("(operator_id not in @df_group.index)&(is_missed_call == False)&(internal == True)")['calls_count']
external = df.query("(operator_id not in @df_group.index)&(is_missed_call == False)&(internal == False)")['calls_count']
print(internal.shape[0], external.shape[0])
```

Проверим выборки на нормальность распределения:

```python
for_shapiro(internal, 'internal')
```

```python
for_shapiro(external, 'external')
```

Выборки не имеют нормального распределения, поэтому лучше использовать параметрический тест Манн-Уитни.

```python
for_test(internal, external)
```

p-value еще меньше, чем в первом тесте, однозначно не можем отвергнуть нулевую гипотезу, а значит, не можем утверждать, что среднее количество отвеченных внешних и внутренних вызовов различно.
